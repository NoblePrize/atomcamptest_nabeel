```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib notebook
```

```python
x_data = np.random.random(1000) * 100
y_data = np.random.random(1000) * 100

plt.scatter (x_data, y_data, c="#00f", marker="*", s=100, alpha=0.3)
```

```python
years = [2006 + x for x in range(16)]
weights = [80,83,84,85,86,82,81,79,83,80,
           82,82,83,81,80,79]

plt.plot(years, weights, "g--", lw=3)
```

```python
x= ["C++", "C#", "Python", "Java", "GO"]
y = [20, 50, 140, 1, 45]

plt.bar (x,y, color="r", align="edge", width=0.5, edgecolor="green", lw=6)
```

```python
ages = np.random.normal(20,1.5,1000)
#print (ages)
#plt.hist(ages,  bins=[ages.min(), 18, 21, ages.max()])
plt.hist(ages,
         bins=20,
         cumulative=True)
```

```python
langs = ["Python", "C++", "Java", "C#", "Go"]
votes = [50, 24, 14, 6, 17]
explodes = [0,0,0,0.2,0]
plt.pie(votes, labels=langs, explode = explodes, autopct="%.2f%%", pctdistance=1.3, startangle=90)
```

```python
heights = np.random.normal(172,8,300)
plt.boxplot(heights)
```

```python
first = np.linspace (0, 10, 25)
second = np.linspace (10, 200, 25)
third = np.linspace (200, 210, 25)
fourth = np.linspace (210, 230, 25)

data = np.concatenate((first, second, third, fourth))
plt.boxplot(data)
```

```python
years = [2014 + x for x in range(8)]
income = [55, 56, 62, 61,
         72, 72, 73,75]

income_ticks = list(range(50,81,2))
plt.plot(years, income)
plt.title("Income of John (in USD)", fontsize=20, fontname="Calibri")
plt.xlabel("Year")
plt.ylabel("Yearly Income in USD")
plt.yticks(income_ticks, [f"${x}k" for x in income_ticks])
plt.show()
```
